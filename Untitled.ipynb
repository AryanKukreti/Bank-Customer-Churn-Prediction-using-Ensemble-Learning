## Bank Customer Churn Prediction 

```python
#importing all the libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme(color_codes = True)
```

```python
df = pd.read_csv('Bank Customer Churn Prediction.csv')
df.head(5)
```

**We are gonna check if there exists any null values within**

```python
df.isnull().head()
```

```python
df.isnull().sum()
```

```python
# we are going to count 1's and 0's in our churn column
color_wheel = {1: "#0392cf", 2: "#07bc043"}
colors = df['churn'].map(lambda x: color_wheel.get(x+1))
print(df.churn.value_counts())
p = df.churn.value_counts().plot(kind="bar")
```

```python
df['country'] = df['country'].replace(['Germany'], '0')
df['country'] = df['country'].replace(['France'], '1')
df['country'] = df['country'].replace(['Spain'], '2')

df['gender'] = df['gender'].replace(['Male'], '1')
df['gender'] = df['gender'].replace(['Female'], '0')
```

```python
df.head(5)
```

```python
#we are changing our object type to integer otherwise they cannot be used further without changing them

df['country'] = pd.to_numeric(df['country'])
df['gender'] = pd.to_numeric(df['gender'])
```

```python
#next we are going to drop customer id as it has no use within our prediction
df = df.drop(['customer_id'], axis='columns')
df.head()
```

# Let's have a look at the heatmap to get an idea

```python
sns.heatmap(df.corr(), fmt = '0.2g')
```

```python
from sklearn.model_selection import train_test_split
# Assuming 'churn' is the target variable
X = df.drop(['churn'], axis=1)
y = df['churn']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## DECISION TREES

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

## XG BOOST

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb.fit(X_train, y_train)
```

## SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC

svm = SVC(probability=True, random_state=42)  # Use probability=True to get prediction probabilities
svm.fit(X_train, y_train)
```

```python
#we are checking the scores and metrics of our problem

from sklearn.metrics import classification_report, roc_auc_score

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    # Print classification report with zero_division parameter
    print(f"{model.__class__.__name__} Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))
    y_prob = model.predict_proba(X_test)[:, 1]
    print(f"{model.__class__.__name__} ROC AUC Score: {roc_auc_score(y_test, y_prob)}\n")

# Evaluate each model
print("Decision trees \n\n")
evaluate_model(dt, X_test, y_test)

print("Random Forest \n\n")
evaluate_model(rf, X_test, y_test)

print("SVM \n\n")
evaluate_model(svm, X_test, y_test)

print("XG Boost \n")
evaluate_model(xgb, X_test, y_test)
```

```python
#here we are using the stacking classifier to get the best from all the models

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

# Defining base models
base_models = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('svm', SVC(probability=True, random_state=42)),
    ('xgb', XGBClassifier(eval_metric='mlogloss'))  # Removed use_label_encoder
]

# Defining the  meta-model
meta_model = LogisticRegression()

# training  the stacking model
stacking_clf = StackingClassifier(estimators=base_models, final_estimator=meta_model)
stacking_clf.fit(X_train, y_train)

# The final step involves evaluation of our model with the test data
evaluate_model(stacking_clf, X_test, y_test)
```
